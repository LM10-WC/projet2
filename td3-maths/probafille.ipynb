```python
import numpy as np
import matplotlib.pyplot as plt

# Données
filles_total = 35
filles_reussies = 18
garcons_total = 25
garcons_reussis = 12

# Calcul de la probabilité conditionnelle P(Fille | Réussite)
total_reussis = filles_reussies + garcons_reussis
probabilite_fille_sachant_reussite = filles_reussies / total_reussis

# Affichage des résultats
print(f"Probabilité qu'un étudiant qui a réussi soit une fille : {probabilite_fille_sachant_reussite:.2%}")

# Création du graphique circulaire (camembert)
labels = ["Filles ayant réussi", "Garçons ayant réussi"]
sizes = [filles_reussies, garcons_reussis]
colors = ["#4CAF50", "#FF5252"]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, startangle=90, wedgeprops={"edgecolor": "black"})
ax.set_title("Répartition des étudiants ayant réussi")

# Interprétation automatique
if probabilite_fille_sachant_reussite > 0.5:
    interpretation = "La majorité des étudiants qui ont réussi sont des filles."
elif probabilite_fille_sachant_reussite == 0.5:
    interpretation = "Il y a une répartition équilibrée entre filles et garçons parmi ceux qui ont réussi."
else:
    interpretation = "La majorité des étudiants qui ont réussi sont des garçons."

plt.figtext(0.5, 0.01, interpretation, wrap=True, horizontalalignment='center', fontsize=10, color="black")

plt.show()
```
