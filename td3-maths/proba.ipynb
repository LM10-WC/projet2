```python
import numpy as np
import matplotlib.pyplot as plt

# Données
total_etudiants = 60
filles_reussies = 18
garcons_reussis = 12

# Calcul de la probabilité de réussite
total_reussis = filles_reussies + garcons_reussis
probabilite_reussite = total_reussis / total_etudiants

# Définition du commentaire basé sur la probabilité
if probabilite_reussite >= 0.75:
    interpretation = "Le taux de réussite est élevé, ce qui indique que la majorité des étudiants ont bien performé."
elif 0.5 <= probabilite_reussite < 0.75:
    interpretation = "Le taux de réussite est modéré, montrant une répartition équilibrée entre réussites et échecs."
else:
    interpretation = "Le taux de réussite est relativement bas, suggérant des difficultés générales dans le test."

# Affichage de la probabilité et de l'interprétation
print(f"Probabilité qu'un étudiant ait réussi : {probabilite_reussite:.2%}")
print(f"Interprétation : {interpretation}")

# Création du graphique circulaire (camembert)
labels = ["Réussite", "Échec"]
sizes = [total_reussis, total_etudiants - total_reussis]
colors = ["#4CAF50", "#FF5252"]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, startangle=90, wedgeprops={"edgecolor": "black"})
ax.set_title("Probabilité de réussite d'un étudiant")

# Ajout d'un texte d'interprétation sur le graphique
plt.figtext(0.5, 0.01, interpretation, wrap=True, horizontalalignment='center', fontsize=10, color="black")

plt.show()
```
