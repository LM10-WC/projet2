```python
import numpy as np
import matplotlib.pyplot as plt

# Création de deux matrices 4x4 aléatoires
np.random.seed(42)  # Pour reproductibilité
matrice_A = np.random.randint(1, 10, (4, 4))
matrice_B = np.random.randint(1, 10, (4, 4))

# Calculs des résultats
somme_matrices = matrice_A + matrice_B
produit_element_par_element = matrice_A * matrice_B
produit_matriciel = np.dot(matrice_A, matrice_B)

# Définir les limites de couleurs pour une échelle cohérente
vmin = min(somme_matrices.min(), produit_element_par_element.min(), produit_matriciel.min())
vmax = max(somme_matrices.max(), produit_element_par_element.max(), produit_matriciel.max())

# Affichage des résultats
print("Matrice A:\n", matrice_A)
print("Matrice B:\n", matrice_B)
print("\nSomme des matrices:\n", somme_matrices)
print("\nProduit élément par élément:\n", produit_element_par_element)
print("\nProduit matriciel:\n", produit_matriciel)

# Visualisation des matrices
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

matrices = [somme_matrices, produit_element_par_element, produit_matriciel]
titres = ["Somme des matrices", "Produit élément par élément", "Produit matriciel"]

for ax, matrice, titre in zip(axes, matrices, titres):
    im = ax.imshow(matrice, cmap="coolwarm", vmin=vmin, vmax=vmax)
    ax.set_title(titre)
    ax.axis("off")

    # Ajout des valeurs sur les matrices
    for i in range(matrice.shape[0]):
        for j in range(matrice.shape[1]):
            ax.text(j, i, str(matrice[i, j]), ha='center', va='center', color='black')

# Ajout de la barre de couleurs
fig.colorbar(im, ax=axes, orientation='vertical', fraction=0.05)

plt.tight_layout()
plt.show()
```
