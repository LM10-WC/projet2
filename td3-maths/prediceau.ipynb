```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Donnes simules
taux_pauvrete = np.array([20, 30, 40, 50, 60, 70, 80, 90, 100]).reshape(-1, 1)
acces_eau_potable = np.array([95, 90, 85, 80, 75, 70, 65, 60, 55])

# Cration du modle de rgression linaire
modele = LinearRegression()
modele.fit(taux_pauvrete, acces_eau_potable)

# Prdictions sur les donnes
predictions = modele.predict(taux_pauvrete)

# Extraction des paramtres du modle
pente = modele.coef_[0]
intercept = modele.intercept_

# Affichage des rsultats
print(f"Pente (coefficient) : {pente:.4f}")
print(f"Intercept : {intercept:.4f}")

# Visualisation des rsultats
plt.figure(figsize=(8, 5))
plt.scatter(taux_pauvrete, acces_eau_potable, color="blue", label="Donnes relles")
plt.plot(taux_pauvrete, predictions, color="red", linestyle="--", label="Rgression linaire")
plt.xlabel("Taux de pauvret (%)")
plt.ylabel("Accs  l'eau potable (%)")
plt.title("Prdiction de l'accs  l'eau potable en fonction du taux de pauvret")
plt.legend()
plt.grid()

# Ajout de l'interprtation sur le graphique
interpretation = f"Pente : {pente:.4f}\nIntercept : {intercept:.4f}\nUne augmentation de 1% du taux de pauvret entrane en moyenne une baisse de {abs(pente):.2f}% de l'accs  l'eau potable."
plt.figtext(0.5, 0.01, interpretation, wrap=True, horizontalalignment='center', fontsize=10, color="black")

plt.show()
```
