```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Données simulées
taux_pauvrete = np.array([20, 30, 40, 50, 60, 70, 80, 90, 100]).reshape(-1, 1)
acces_eau_potable = np.array([95, 90, 85, 80, 75, 70, 65, 60, 55])

# Création du modèle de régression linéaire
modele = LinearRegression()
modele.fit(taux_pauvrete, acces_eau_potable)

# Prédictions sur les données
predictions = modele.predict(taux_pauvrete)

# Extraction des paramètres du modèle
pente = modele.coef_[0]
intercept = modele.intercept_

# Affichage des résultats
print(f"Pente (coefficient) : {pente:.4f}")
print(f"Intercept : {intercept:.4f}")

# Visualisation des résultats
plt.figure(figsize=(8, 5))
plt.scatter(taux_pauvrete, acces_eau_potable, color="blue", label="Données réelles")
plt.plot(taux_pauvrete, predictions, color="red", linestyle="--", label="Régression linéaire")
plt.xlabel("Taux de pauvreté (%)")
plt.ylabel("Accès à l'eau potable (%)")
plt.title("Prédiction de l'accès à l'eau potable en fonction du taux de pauvreté")
plt.legend()
plt.grid()

# Ajout de l'interprétation sur le graphique
interpretation = f"Pente : {pente:.4f}\nIntercept : {intercept:.4f}\nUne augmentation de 1% du taux de pauvreté entraîne en moyenne une baisse de {abs(pente):.2f}% de l'accès à l'eau potable."
plt.figtext(0.5, 0.01, interpretation, wrap=True, horizontalalignment='center', fontsize=10, color="black")

plt.show()
```
