```python
# import matplotlib.pyplot as plt
import numpy as np

# Demande du choix du solide
print("Choisissez un solide parmi :")
print("1. Sphère\n2. Prisme Rectangulaire\n3. Cône")
choix = int(input("Entrez le numéro correspondant : "))

# Initialisation des volumes
volumes = {"Sphère": 0, "Prisme Rectangulaire": 0, "Cône": 0}

# Calcul du volume selon le choix de l'utilisateur
if choix == 1:
    r = float(input("Entrez le rayon de la sphère : "))
    volumes["Sphère"] = (4/3) * np.pi * r**3

elif choix == 2:
    L = float(input("Entrez la longueur du prisme : "))
    l = float(input("Entrez la largeur du prisme : "))
    h = float(input("Entrez la hauteur du prisme : "))
    volumes["Prisme Rectangulaire"] = L * l * h

elif choix == 3:
    r = float(input("Entrez le rayon de la base du cône : "))
    h = float(input("Entrez la hauteur du cône : "))
    volumes["Cône"] = (1/3) * np.pi * r**2 * h

else:
    print("Choix invalide. Veuillez relancer le programme.")

# Comparaison avec des volumes équivalents
if choix in [1, 2, 3]:
    # Affichage du volume choisi
    print(f"\nVolume du solide choisi : {volumes[list(volumes.keys())[choix-1]]:.2f} unités³")

    # Création du graphe en barres
    plt.bar(volumes.keys(), volumes.values(), color=['b', 'g', 'r'])
    plt.xlabel("Solides")
    plt.ylabel("Volume (unités³)")
    plt.title("Comparaison des volumes des solides")
    plt.show()
```
