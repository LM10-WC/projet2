```python
import matplotlib.pyplot as plt

# Saisie des ventes pour 7 jours
ventes = []
jours = ["Lundi", "Mardi", "Mercredi", "Jeudi", "Vendredi", "Samedi", "Dimanche"]

for jour in jours:
    vente = float(input(f"Entrez les ventes pour {jour} : "))
    ventes.append(vente)

# Calcul du total et de la moyenne
total_ventes = sum(ventes)
moyenne_ventes = total_ventes / len(ventes)

# Affichage des résultats
print(f"\nTotal des ventes : {total_ventes}")
print(f"Moyenne des ventes : {moyenne_ventes:.2f}")

# Création du graphe
plt.plot(jours, ventes, marker='o', linestyle='-', color='b')
plt.xlabel("Jours de la semaine")
plt.ylabel("Ventes")
plt.title("Évolution des ventes sur 7 jours")
plt.grid(True)
plt.show()
```
