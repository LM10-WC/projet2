```python
import matplotlib.pyplot as plt

# Saisie des températures pour 7 jours
temperatures = []
jours = ["Lundi", "Mardi", "Mercredi", "Jeudi", "Vendredi", "Samedi", "Dimanche"]

for jour in jours:
    temp = float(input(f"Entrez la température pour {jour} : "))
    temperatures.append(temp)

# Calcul de la moyenne des températures
moyenne_temp = sum(temperatures) / len(temperatures)

# Affichage des résultats
print(f"\nTempérature moyenne : {moyenne_temp:.2f}°C")

# Création du graphe
plt.plot(jours, temperatures, marker='o', linestyle='-', color='r')
plt.xlabel("Jours de la semaine")
plt.ylabel("Température (°C)")
plt.title("Évolution de la température sur 7 jours")
plt.grid(True)
plt.show()
```
