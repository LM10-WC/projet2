```python
import matplotlib.pyplot as plt

# Demande du choix de la figure
print("Choisissez une figure parmi :")
print("1. Carré\n2. Losange\n3. Trapèze")
choix = int(input("Entrez le numéro correspondant : "))

# Initialisation des valeurs
surfaces = {"Carré": 0, "Losange": 0, "Trapèze": 0}
perimetres = {}

# Calculs selon le choix de l'utilisateur
if choix == 1:
    c = float(input("Entrez la longueur du côté du carré : "))
    perimetres["Carré"] = 4 * c
    surfaces["Carré"] = c * c

elif choix == 2:
    c = float(input("Entrez la longueur du côté du losange : "))
    D = float(input("Entrez la grande diagonale du losange : "))
    d = float(input("Entrez la petite diagonale du losange : "))
    perimetres["Losange"] = 4 * c
    surfaces["Losange"] = (D * d) / 2

elif choix == 3:
    a = float(input("Entrez la longueur du côté a du trapèze : "))
    b = float(input("Entrez la longueur du côté b du trapèze : "))
    c = float(input("Entrez la longueur du côté c du trapèze : "))
    d = float(input("Entrez la longueur du côté d du trapèze : "))
    h = float(input("Entrez la hauteur du trapèze : "))
    perimetres["Trapèze"] = a + b + c + d
    surfaces["Trapèze"] = ((a + b) * h) / 2

else:
    print("Choix invalide. Veuillez relancer le programme.")

# Affichage des résultats et comparaison
if choix in [1, 2, 3]:
    figure = list(surfaces.keys())[choix - 1]
    print(f"\nPérimètre du {figure} : {perimetres[figure]:.2f} unités")
    print(f"Surface du {figure} : {surfaces[figure]:.2f} unités²")

    # Création du graphe en barres comparant les surfaces
    plt.bar(surfaces.keys(), surfaces.values(), color=['b', 'g', 'r'])
    plt.xlabel("Figures")
    plt.ylabel("Surface (unités²)")
    plt.title("Comparaison des surfaces des figures géométriques")
    plt.show()
```
